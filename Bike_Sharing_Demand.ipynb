##### Observe having a feature with object type, that is (datetime). That needs to be formatted by
#####            1. convert datetime from object type to datetime

# Handle outliers

```python
def handle_outliers(feature):
    upper_lim = fileData[feature].quantile(.95)
    lower_lim = fileData[feature].quantile(.05)

    return fileData[(fileData[feature] < upper_lim) & (fileData[feature] > lower_lim)]
    
fileData = handle_outliers('casual')
fileData = handle_outliers('count')
```
